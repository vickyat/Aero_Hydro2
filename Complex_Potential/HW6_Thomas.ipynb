# 1

$$W = \phi + i \psi$$
$$\phi_{doublet} = \frac{-k}{2 \pi} \frac {cos\theta}{r}$$

$$\psi_{doublet} = \frac{k}{2 \pi} \frac{sin \theta}{r}$$

$$ W = \frac {-k}{2 \pi} \frac{cos\theta}{r} + i \frac{k}{2 \pi} \frac {sin \theta}{r}$$

$$ W = \frac{-k}{2\pi}\frac{1}{r} (cos\theta - isin\theta)$$

$$ cos\theta - isin\theta = e^{-i\theta}$$

$$ W = \frac {-k}{2 \pi r} \frac {1}{re^{i\theta}}$$

$$ z = re^{i\theta}$$

$$ W = \frac{-k}{2\pi} \frac{1}{z}$$

```python
def get_stream_doublet(strength,xd,yd,X,Y):
    psi = -strength/(2*numpy.pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
    return psi

def get_vel_doublet(strength,xd,yd,X,Y):
    u = -strength/(2*numpy.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = -strength/(2*numpy.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    return u,v
```

```python
N = 50
x_start,x_end = -2.0,2.0
y_start,y_end = -1.0,1.0
x=numpy.linspace(x_start,x_end,N)
y=numpy.linspace(y_start,y_end,N)
X,Y=numpy.meshgrid(x,y)
kappa = -1.0
x_doublet,y_doublet = 0.0,0.0
```

```python
u_doublet, v_doublet = get_vel_doublet(kappa,x_doublet,y_doublet,X,Y)
psi_doublet = get_stream_doublet(kappa,x_doublet,y_doublet,X,Y)
```

```python
size =10
pyplot.figure(figsize=(size,(y_end-y_start)/(x_end-x_start)*size))
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,u_doublet,v_doublet,density=2, linewidth=1)
pyplot.scatter(x_doublet,y_doublet, color='#990000',s=80)
pyplot.contour(X,Y,psi_doublet,levels=[0.],colors='#990000',linewidths=2,linestyles='solid');
```

# 2a

$$ W = z^{n} = (x+iy)^{n} \ $$

n = 2:

$$ x^{2}+2iyx + i^{2}y^{2}$$

$$ x^{2}+2iyx - y^{2}$$

--> This will have 2 roots in x and 2 roots in y => 2*2 = 4 total

n = 3:

$$(x^{2}+2iyx-y^{2})(x+iy)$$

$$ x^{3} + 2iyx^{2} -xy^{2} + x^{2}iy + 2i^{2}y{2}x - iy^{3}$$

--> Again, 3 root for x, 3 in y => 2* 3 = 6 total, and so on as you increase the exponent

```python
z = X + 1j*Y

def bigW (n):
    W = z**n
    return W
def littlew(n):
    w = n*z**(n-1)
    return w
```

```python
w = littlew(7)
W = bigW(7)
size =10
pyplot.figure(figsize=(size,(y_end-y_start)/(x_end-x_start)*size))
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,w.real,-w.imag,density=2, linewidth=1)
pyplot.contour(X,Y,W.imag,levels=[0.],colors='#990000',linewidths=2,linestyles='solid');
```

With n = 7, the flow is divided into 2n parts, or 14 parts.

# 3a

$$ \psi_{vortex} = \frac {-\Gamma}{2\pi} ln(r)$$

$$ \phi_{vortex} = \frac{\Gamma}{2\pi}\ \theta$$

$$ W = \frac{\Gamma}{2\pi} \theta - i\ \frac{\Gamma}{2\pi}\ ln(r)$$

$$ W = \frac{-\Gamma}{2\pi}(i\ln(r)\ -\ \theta) $$

$$W =\frac {i^{2}}{i^{2}} \frac{\Gamma}{2\pi}\ (\frac {\theta}{i} - ln(r))$$

$$ W = -i \frac{\Gamma}{2\pi} (ln(r) + i\theta)$$

$$ W = -i \frac{\Gamma}{2\pi} ln(z)$$

$$ w = W' = -i \frac{\Gamma}{2\pi} \frac{1}{z}$$

CW vortex: $$w = -i \frac{\Gamma}{2\pi} \frac{1}{z+a}$$

CCW vortex: $$w = i \frac{\Gamma}{2\pi} \frac{1}{z-a}$$

$$ w = i \frac{\Gamma}{2\pi} \frac{2a}{z^{2}-a^{2}}$$

```python
gamma1 = 5.0
x_vortex1,y_vortex1 = 1.0,0.0
gamma2 = -5.0
x_vortex2,y_vortex2 = -1.0,0.0

def get_vel_vortex(strength,xv,yv,X,Y):
    u=+strength/(2*numpy.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v=-strength/(2*numpy.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u,v

def get_stream_vortex(strength,xv,yv,X,Y):
    psi = strength/(4*numpy.pi)*numpy.log((X-xv)**2+(Y-yv)**2)
    return psi
```

```python
u_vortex1,v_vortex1=get_vel_vortex(gamma1,x_vortex1,y_vortex1,X,Y)
psi_vortex1=get_stream_vortex(gamma1,x_vortex1,y_vortex1,X,Y)

u_vortex2,v_vortex2=get_vel_vortex(gamma2,x_vortex2,y_vortex2,X,Y)
psi_vortex2=get_stream_vortex(gamma2,x_vortex2,y_vortex2,X,Y)
```

```python
u = u_vortex1+u_vortex2
v = v_vortex1+v_vortex2
psi=psi_vortex1+psi_vortex2
```

```python
def jt1(c,T,H,R,THETA):
    zetac = R*numpy.exp(1j*THETA)
    zeta0 = c*(-4/(3*numpy.sqrt(3))*T + (1j*2*H))
    zeta = zetac + zeta0
    z = zeta + (c**2/zeta)
    return z,zeta
```
