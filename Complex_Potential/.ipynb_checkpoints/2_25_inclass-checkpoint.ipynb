```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
def jt1(a,T,H,R,THETA):
    zetac = R*numpy.exp(1j*THETA)
    zeta0 = a*(-4/(3*numpy.sqrt(3))*T + (1j*2*H))
    zeta = zetac + zeta0
    z = zeta + (a**2/zeta)
    return z
```

```python
def get_vel_cyl(u_inf,R,THETA):
    v_r = u_inf*numpy.cos(THETA)*(1-(r**2/(R)**2))
    v_theta = -u_inf*numpy.sin(THETA)*(1+(r**2/(R)**2))
    return v_r,v_theta

def get_stream_cyl(u_inf,R,THETA):
    psi = u_inf*(R)*numpy.sin(THETA)*(1-(r**2/(R)**2))
    return psi
```

```python
T =0.1
c=1
a = c/4
H = 0.04
rc = (0.25 + T/(3*numpy.sqrt(3)))*c
r = numpy.linspace(rc,2*rc,15)
theta = numpy.linspace(0,2*numpy.pi,50)
R,THETA = numpy.meshgrid(r,theta)
```

```python
z,zeta = jt1(a,T,H,R,THETA)
```
