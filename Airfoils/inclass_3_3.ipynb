```python
N = 100
x_start,x_end = -1.5,1.5
y_start,y_end = -1.5,1.5
x=numpy.linspace(x_start,x_end,N)
y=numpy.linspace(y_start,y_end,N)
X,Y = numpy.meshgrid(x,y)
```

```python
u_inf = 1.0
u_fs = u_inf*numpy.ones((N,N),dtype=float)
v_fs = numpy.zeros((N,N),dtype=float)
psi_fs = u_inf*Y*numpy.ones((N,N),dtype=float)
```

```python
sigma = 2*u_inf
y_min,y_max = -1.0,1.0

u_sheet = numpy.empty((N,N),dtype=float)
v_sheet = numpy.empty((N,N),dtype=float)
for i in range(N):
    for j in range(N):
        integrand = lambda s: X[i,j]/(X[i,j]**2+(Y[i,j]-s)**2)
        u_sheet[i,j] = sigma/(2*numpy.pi)*integrate.quad(integrand,y_min,y_max)[0]
        
        integrand=lambda s: (Y[i,j]-s)/(X[i,j]**2+(Y[i,j]-s)**2)
        v_sheet[i,j] = sigma/(2*numpy.pi)*integrate.quad(integrand,y_min,y_max)[0]
        
u=u_fs+u_sheet
v=v_fs+v_sheet
```

```python
pyplot.figure(figsize=(7,7))
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,u,v,density=2)
pyplot.axvline(0.0,(y_min-y_start)/(y_end-y_start), (y_max-y_start)/(y_end-y_start),\
               color='r',linewidth=4)
velocity=pyplot.contourf(X,Y,numpy.sqrt(u**2+v**2), levels = numpy.linspace(0.0,0.1,10))
cbar = pyplot.colorbar(velocity,ticks=[0,0.05,0.1], orientation='horizontal')
cbar.set_label('vel mag')
pyplot.show()
```

```python
def get_sigma(s):
    if  s<= 0.5:
        sigma = 1
    else:
        sigma = -1
    return sigma
```
