```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
class Panel:
    def __init__(self,xa,ya,xb,yb):
        self.xa,self.ya = xa,ya
        self.xb,self.yb = xb,yb
        
        self.xc,self.yc = (xa+xb)/2, (ya+yb)/2
        self.length = numpy.sqrt((xb-xa)**2+(yb-ya)**2)
        if xb-xa <= 0.:
            self.beta = numpy.arccos((yb-ya)/self.length)
        elif xb-xa > 0.:
            self.beta = numpy.pi + numpy.arccos(-(yb-ya)/self.length)
            
        self.n=[numpy.cos(self.beta),numpy.sin(self.beta)]
        self.t=[-numpy.sin(self.beta),numpy.cos(self.beta)]
        self.sigma = 1
        self.vt = 0
        self.cp = 0
```

```python
class FS:
    def __init__(self,u_inf,alpha):
        self.u_inf = u_inf
        self.alpha = numpy.radians(alpha)
```

```python
u_inf = 1
alpha = 0.0
fs = FS(u_inf,alpha)
```

```python
def panel_vel(p,x,y):
    x_p = numpy.sin(p.beta)*(x-p.xb) - numpy.cos(p.beta)*(y-p.yb)
    y_p = numpy.cos(p.beta)*(x-p.xb) + numpy.sin(p.beta)*(y-p.yb)
    u_p = p.sigma/(4*numpy.pi)*numpy.log((x_p**2+y_p**2)/((x_p-p.length)**2+y_p**2))
    v_p = p.sigma/(2*numpy.pi)*(numpy.arctan2(y_p,x_p-p.length)-numpy.arctan2(y_p,x_p))
    u = numpy.sin(p.beta)*u_p + numpy.cos(p.beta)*v_p
    v = -numpy.cos(p.beta)*u_p + numpy.sin(p.beta)*v_p
    return u,v
```

```python
# create a single constant-source panel
N_panels=20
panels = numpy.empty(N_panels, dtype=object)
x_ends = numpy.array([0.0, 0.5])
y_ends = numpy.array([0.01, 0.1])
panels[0] = Panel(x_ends[0], y_ends[0], x_ends[1], y_ends[1])

# create mesh grid
N = 81                                # number of points in each direction
x_start, x_end = -1.0, 1.5            # boundaries in the x-direction
y_start, y_end = -1.0, 1.5            # boundaries in the y-direction
x = numpy.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = numpy.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid

# compute velocity field
u = numpy.empty((N, N), dtype=float)
v = numpy.empty((N, N), dtype=float)
for i in range(N):
    for j in range(N):  
        u[i,j], v[i,j] = panel_vel(panels[0],X[i,j],Y[i,j])


pyplot.figure(figsize=(10, 10))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.plot(x_ends, y_ends, color='#CD2305', linestyle='-', linewidth=2)
pyplot.streamplot(X, Y, u, v, density=1, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end);
```

```python
a = 1/4
T = .10
H = 0
R1=0.25 + T/(3*numpy.sqrt(3))
x_ends=R1*numpy.cos(numpy.linspace(0,2*numpy.pi,N_panels+1))
y_ends = R1*numpy.sin(numpy.linspace(0,2*numpy.pi,N_panels+1))
zeta = x_ends + 1j*y_ends
zeta0 = -4*a*T/(3*numpy.sqrt(3))+1j*2*a*H

z = (zeta0+zeta) + a**2/(zeta0+zeta)
for i in range(N_panels):
    panels[i] = Panel(z.real[i],z.imag[i],z.real[i+1],z.imag[i+1])
```

```python
def build_matrix(panels):
    N=len(panels)
    A = numpy.empty((N,N),dtype=float)
    numpy.fill_diagonal(A,0.5)
    
    for i, p_i in enumerate(panels):
        for j,p_j in enumerate(panels):
            if i != j:
                A[i,j] = 0.5/numpy.pi*numpy.sqrt(((panel_vel(panels[j],x,y))[0])**2 + ((panel_vel(panels[j],x,y))[1])**2)
    return A

def build_rhs(panels,fs):
    b = numpy.empty(len(panels),dtype=float)
    for i, panel in enumerate(panels):
        b[i] = -fs.u_inf * numpy.cos(fs.alpha - panel.beta)
    return b
```

```python
A = build_matrix(panels)
b = build_rhs(panels,fs)
```

```python
sigma = numpy.linalg.solve(A,b)
for i, panel in enumerate(panels):
    panel.sigma=sigma[i]
```
