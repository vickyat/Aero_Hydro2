```python
import numpy
from matplotlib import pyplot
from matplotlib import rcParams
%matplotlib inline
rcParams['font.size']=16
```

```python
def jt1(a,T,H,R,THETA):
    zetac = R*numpy.exp(1j*THETA)
    zeta0 = a*(-4/(3*numpy.sqrt(3))*T + (1j*2*H))
    zeta = zetac + zeta0
    z = zeta + (a**2/zeta)
    return z
```

```python
Ta,Ha = 0.05,0.0
Tb,Hb = 0.1,0.0
Tc,Hc = 0.3,0.0
Td,Hd = 0.1,0.02
Te,He = 0.1,0.05
Tf,Hf = 0.1,0.1
```

```python
c = 1
a = c/4
def grid(T):
    rc = (0.25+T/(3*numpy.sqrt(3)))*c
    r=numpy.linspace(rc,3*rc,15)
    theta = numpy.linspace(0,2*numpy.pi,50)
    R,THETA = numpy.meshgrid(r,theta)
    return rc,r,theta,R,THETA
```

# a

```python
rca,ra,thetaa,Ra,THETAa = grid(Ta)
z_a = jt1(a,Ta,Ha,Ra,THETAa)

pyplot.scatter(z_a.real,z_a.imag)
```

# b

```python
rcb,rb,thetab,Rb,THETAb = grid(Tb)
z_b = jt1(a,Tb,Hb,Rb,THETAb)
pyplot.scatter(z_b.real,z_b.imag)
```

# c

```python
rcc,rc,thetac,Rc,THETAc = grid(Tc)
z_c = jt1(a,Tc,Hc,Rc,THETAc)
print(rcc)
pyplot.scatter(z_c.real,z_c.imag)
```

```python
rce,re,thetae,Re,THETAe = grid(Te)
z_e = jt1(a,Te,He,Re,THETAe)
pyplot.scatter(z_e.real,z_e.imag)
```

```python
# plot streamlines of af d alpha = 0 deg
u_infd = 1
alphad0 = numpy.radians(0)
gammad0 = 4*numpy.pi*rcd*u_infd*numpy.sin(alphad0+(2*Hd))
kappad0 = rcd**2*(u_infd*2*numpy.pi)

zeta0d0 = a*(-4/(3*numpy.sqrt(3))*Td+ (1j*2*Hd))
zetafd0 = (Rd)*numpy.exp(1j*(THETAd))
zetad0 = (zetafd0-zeta0d0)*numpy.exp(-1j*alphad0)#zetafd - zeta0d
zd0 = zetad0+ a**2/zetad0
Wd0 = u_infd*(zetad0+(rcd**2/zetad0))+ 1j*gammad0*numpy.log(zetad0/rcd)/(2*numpy.pi)
Wdz0 = u_infd*(zd0+(rcd**2/zd0))+ 1j*gammad0*numpy.log(zd0/rcd)/(2*numpy.pi)
```

```python
pyplot.figure(figsize=(7,7))
pyplot.contour(zetafd0.real,zetafd0.imag,Wd0.imag,50,linewidths=3);
```

```python
pyplot.figure(figsize=(7,7))
#pyplot.scatter(z_d.real,z_d.imag)
pyplot.contour(z_d.real,z_d.imag,Wd0.imag,100,linewidths = 3);
```

```python
# cp along af d, alpha=0
#did not normalize x axis since c=1 and "x-values" are in terms of R and Theta
wzetad0 = u_infd - (u_infd*rcd**2/zetad0**2) + 1j*gammad0/(numpy.pi*2*zetad0)
wzd0 = wzetad0/(1-(a**2/zetad0**2))

cpd0 = 1-(wzd0[:25,0]**2/u_infd**2)
cpd00 =  1-(wzd0[25:,0]**2/u_infd**2)

pyplot.figure(figsize=(7,7))
pyplot.gca().invert_yaxis()
pyplot.plot(z_d.real[:25,0],cpd0, label='upper')
pyplot.plot(z_d.real[25:,0],cpd00, 'r',label='lower')
pyplot.legend(loc=2,prop={'size':10})
pyplot.title('$C_{p}$ along airfoil surface')
pyplot.xlabel('position along chord')
pyplot.ylabel('$C_{p}$')
pyplot.show()
```
