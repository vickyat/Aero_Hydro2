```python
import numpy
import math
from matplotlib import pyplot
%matplotlib inline
```

```python
N = 100
x_start,x_end=-1.0,1.0
y_start,y_end=-1.5,1.5
x=numpy.linspace(x_start,x_end,N)
y=numpy.linspace(y_start,y_end,N)
X,Y=numpy.meshgrid(x,y)
```

```python
u_inf = 1.0
u_fs = u_inf*numpy.ones((N,N),dtype=float)
v_fs = numpy.zeros((N,N),dtype=float)
```

```python
class Source:
    def __init__(self,strength,x,y):
        self.strength=strength
        self.x,self.y=x,y
    
    def velocity(self,X,Y):
        self.u=self.strength/(2*numpy.pi)*(X-self.x)/((X-self.x)**2+(Y-self.y)**2)
        self.v=self.strength/(2*numpy.pi)*(Y-self.y)/((X-self.x)**2+(Y-self.y)**2)
    
    def stream(self,X,Y):
        self.psi = self.strength/(2*numpy.pi)*numpy.arctan2((Y-self.y),(X-self.x))
```

```python
N_sources=11
strength=5.0
strength_source=strength/N_sources
x_source=numpy.zeros(N_sources,dtype=float)
y_source=numpy.linspace(-1.0,1.0,N_sources)

sources=numpy.empty(N_sources,dtype=object)
for i in range(N_sources):
    sources[i]=Source(strength_source,x_source[i],y_source[i])
    sources[i].velocity(X,Y)
    
u = u_fs.copy()
v = v_fs.copy()
for source in sources:
    u+=source.u
    v+=source.v
```

```python
pyplot.figure(figsize=(6,7))
pyplot.grid(True)
pyplot.streamplot(X,Y,u,v,density=2)
pyplot.scatter(x_source,y_source,color='r',s=80,marker='o')
vel = pyplot.contourf(X,Y,numpy.sqrt(u**2+v**2),levels=numpy.linspace(0.0,0.1,10))
cbar = pyplot.colorbar(vel,ticks=[0,0.05,0.1],orientation='horizontal')
cbar.set_label('vel mag',fontsize=16);
```

```python
sigma =2*u_inf
y_min,y_max=-1.0,1.0
u_sheet=numpy.empty((N,N),dtype=float)
v_sheet=numpy.empty((N,N),dtype=float)
for i in range(N):
    for j in range(N):
        integrand=lambda s: X[i,j]/(X[i,j]**2+(Y[i,j]-s)**2)
        u_sheet[i,j]=sigma/(2*numpy.pi)*integrate.quad(integrand,y_min,y_max)[0]
        
        integrand=lambda s: (Y[i,j]-s)/(X[i,j]**2+(Y[i,j]-s)**2)
        v_sheet[i,j]=sigma/(2*numpy.pi)*integrate.quad(integrand,y_min,y_max)[0]
        
u=u_fs+u_sheet
v=v_fs+v_sheet
```

```python
pyplot.streamplot(X,Y,u,v,density=2)
pyplot.axvline(0.0,(y_min-y_start)/(y_end-y_start),(y_max-y_start)/(y_end-y_start),color='r',linewidth=4)
vel = pyplot.contourf(X,Y,numpy.sqrt(u**2+v**2),levels=numpy.linspace(0.0,0.1,10))
cbar = pyplot.colorbar(vel,ticks=[0,0.05,0.1],orientation='horizontal')
cbar.set_label('vel mag');
```
