```python
import numpy
from scipy import integrate
from matplotlib import pyplot
%matplotlib inline
```

```python
u_inf = 1.0
R = 1.0
theta = numpy.linspace(0,2*numpy.pi,100)
```

```python
class Panel:
    def __init__(self,xa,ya,xb,yb):
        self.xa, self.ya= xa,ya
        self.xb, self.yb= xb,yb
        self.xc, self.yc = (xa+xb)/2, (ya+yb)/2 #center pt
        self.length = numpy.sqrt((xb-xa)**2 + (yb-ya)**2)  #len of panel
        
        if xb-xa<=0.:
            self.beta = numpy.arccos((yb-ya)/self.length)
        elif xb-xa >=0.:
            self.beta=numpy.pi+numpy.arccos(-(yb-ya)/self.length)
        
        self.nu, self.nv= numpy.cos(self.beta),numpy.sin(self.beta)
        self.tu, self.tv = numpy.cos(self.beta+(numpy.pi/2)),numpy.sin(self.beta+(numpy.pi/2))
        
        self.sigma = 1
        self.vt = 0
        self.xp = 0
```

```python
N_panels = 20
a = 1/4
T = .10
H = 0

R1=0.25 + T/(3*numpy.sqrt(3))
x_ends=R1*numpy.cos(numpy.linspace(0,2*numpy.pi,N_panels+1))
y_ends = R1*numpy.sin(numpy.linspace(0,2*numpy.pi,N_panels+1))

zeta = x_ends + 1j*y_ends
zeta0 = -4*a*T/(3*numpy.sqrt(3))+1j*2*a*H

z = (zeta0+zeta) + a**2/(zeta0+zeta)
```

```python
panels_xform = numpy.empty(N_panels,dtype=object)
for i in range(N_panels):
    panels_xform[i] = Panel(z.real[i],z.imag[i],z.real[i+1],z.imag[i+1])
pyplot.figure(figsize=(8,5))
pyplot.grid(True)
pyplot.plot(z.real,z.imag,color='r')
pyplot.scatter([p.xa for p in panels_xform],[p.ya for p in panels_xform],color='r',s=40)
pyplot.ylim(-0.3,0.3)
pyplot.scatter([p.xc for p in panels_xform],[p.yc for p in panels_xform], color='k',s=40, zorder=3,marker='x');
pyplot.quiver([p.xc for p in panels_xform],[p.yc for p in panels_xform],[p.nu for p in panels_xform],[p.nv for p in panels_xform],scale=15)
pyplot.quiver([p.xc for p in panels_xform],[p.yc for p in panels_xform],[p.tu for p in panels_xform],[p.tv for p in panels_xform],scale=15,color='b')
```

```python
def get_source_panel_velocity(pan,x,y):
    X,Y = numpy.meshgrid(x,y)
    nx, ny = X.shape
    u,v=numpy.empty((nx,ny),dtype=float),numpy.empty((nx,ny),dtype=float)
    for i in range(nx):
        for j in range(ny):
            
            #u[i,j] = 1/(4*numpy.pi)*numpy.log(((X[i,j]-0)**2+(Y[i,j]-0.01)**2)/((X[i,j]-0.5)**2+(Y[i,j]-0.1)**2))
            #v[i,j] = 1/(4*numpy.pi)*numpy.arctan2((Y[i,j]-0.1),(X[i,j]-0.5))-numpy.arctan2((Y[i,j]-0.01),(X[i,j]-0))  
    
            
            u[i,j] = 1/(4*numpy.pi)*([numpy.log(((X[i,j]-(p.xb))**2+(Y[i,j]-(p.yb))**2)/\
                    ((X[i,j]-(p.xa))**2+(Y[i,j]-(p.ya))**2)) for p in panels])[pan]
            v[i,j] = 1/(4*numpy.pi)*([(numpy.arctan2((Y[i,j]-(p.ya)),(X[i,j]-(p.xa)))-\
                    numpy.arctan2((Y[i,j]-(p.yb)),(X[i,j]-(p.xb)))) for p in panels])[pan]
    return u,v,X,Y
```
