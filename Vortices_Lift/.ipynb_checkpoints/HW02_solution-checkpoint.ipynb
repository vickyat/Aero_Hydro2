```python
import math
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
# Problem 1
N = 50                                # number of points in each direction
x_start, x_end = -8.0, 8.0            # boundaries in the x-direction
y_start, y_end = -8.0, 8.0            # boundaries in the y-direction
x = numpy.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = numpy.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid

# part a
sig_source = 2.0                           # source strength
sig_sink = -2.0                             # sink strength
x_source, y_source = -2.0, 0.0             # location of the source
x_sink, y_sink = 2.0, 0.0                  # location of the sink

# compute the velocity field on the mesh grid
u_source = sig_source/(2*math.pi) * (X-x_source)/((X-x_source)**2 + (Y-y_source)**2)
v_source = sig_source/(2*math.pi) * (Y-y_source)/((X-x_source)**2 + (Y-y_source)**2)
u_sink = sig_sink/(2*math.pi) * (X-x_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
v_sink = sig_sink/(2*math.pi) * (Y-y_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
u = u_source + u_sink
v = v_source + v_sink

# plot the streamlines
size = 5
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=1, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
pyplot.scatter(x_sink, y_sink, color='#33cc33', s=80, marker='o', linewidth=0);
```

```python
# part b
sig_source = 2.0                           # source strength
sig_sink = -1.0                             # sink strength
x_source, y_source = -2.0, 0.0             # location of the source
x_sink, y_sink = 2.0, 0.0                  # location of the sink

# compute the velocity field on the mesh grid
u_source = sig_source/(2*math.pi) * (X-x_source)/((X-x_source)**2 + (Y-y_source)**2)
v_source = sig_source/(2*math.pi) * (Y-y_source)/((X-x_source)**2 + (Y-y_source)**2)
u_sink = sig_sink/(2*math.pi) * (X-x_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
v_sink = sig_sink/(2*math.pi) * (Y-y_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
u = u_source + u_sink
v = v_source + v_sink

# plot the streamlines
size = 5
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
pyplot.scatter(x_sink, y_sink, color='#33cc33', s=80, marker='o', linewidth=0);
```

```python
# part c
sig_source = 2.0                           # source strength
sig_sink = -0.5                             # sink strength
x_source, y_source = -2.0, 0.0             # location of the source
x_sink, y_sink = 2.0, 0.0                  # location of the sink

# compute the velocity field on the mesh grid
u_source = sig_source/(2*math.pi) * (X-x_source)/((X-x_source)**2 + (Y-y_source)**2)
v_source = sig_source/(2*math.pi) * (Y-y_source)/((X-x_source)**2 + (Y-y_source)**2)
u_sink = sig_sink/(2*math.pi) * (X-x_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
v_sink = sig_sink/(2*math.pi) * (Y-y_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
u = u_source + u_sink
v = v_source + v_sink

# plot the streamlines
size = 5
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=1, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
pyplot.scatter(x_sink, y_sink, color='#33cc33', s=80, marker='o', linewidth=0);
```

As the sink reduces in strength, it absorbs less of the flow from the source.  The difference in stengths is equal to the rate of flow escaping out to infinity (by incompressibility, the flow must go somewhere, so whatever does not get absorbed by the sink goes off to infinity).  Since the flow out from the source is uniform in every direction, the ratio of source to sink strength is the proportion of 360 deg. of stream lines from the source that eventually end up in the sink.

```python
# Problem 2
N = 50                                # number of points in each direction
x_start, x_end = -4.0, 4.0            # boundaries in the x-direction
y_start, y_end = -4.0, 4.0            # boundaries in the y-direction
x = numpy.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = numpy.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid

sig_source = 1.0                          # source strength
x_source, y_source = 0.0, 0.0             # location of the source

# compute the velocity field on the mesh grid
u = sig_source/(2*math.pi) * (X-x_source)/((X-x_source)**2 + (Y-y_source)**2)
v = sig_source/(2*math.pi) * (Y-y_source)/((X-x_source)**2 + (Y-y_source)**2)

# integrate one side of a square using a Riemann sum
u_int = u[:,-1]              # grab last column
dy = y[1]-y[0]               # y spacing of grid
Q = 4*numpy.sum(u_int*dy)    # sum up u*dy to approximate integral, multiply by 4 to get full square
print(Q)
```

```python
# try again using a column closer to the source
u_int = u[11:40,30]          # grab column 30
dy = y[1]-y[0]               # y spacing of grid
Q = 4*numpy.sum(u_int*dy)    # sum up u*dy to approximate integral, multiply by 4 to get full square
print(Q)
```

The approximation to the theoretical value, 1.0, gets worse as the path of integration gets coarser, i.e. less points.  This is because the approximation of the integral as a Riemann sum deviates as a function of $max(dx)$.

```python
# Problem 3
N = 50                                # number of points in each direction
x_start, x_end = -4.0, 4.0            # boundaries in the x-direction
y_start, y_end = -4.0, 4.0            # boundaries in the y-direction
x = numpy.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = numpy.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid

# sources and sinks
sig_source = [2.0, 1.0, 4.0]                           # source strength
sig_sink = [-2.0, -2.0, -1,0]                          # sink strength
x_source, y_source = [-2.0, -1.0, 0.0], [0.0, 1.0, -1.0]            # location of the source
x_sink, y_sink = [2.0, 0.0, 2.0], [0.0, 1.0, -2.5]                  # location of the sink

# compute the velocity field on the mesh grid
u_source = 0*X
v_source = 0*X
u_sink = 0*X
v_sink = 0*X
for i in range(3):
    u_source = u_source + sig_source[i]/(2*math.pi) * (X-x_source[i])/((X-x_source[i])**2 + (Y-y_source[i])**2)
    v_source = v_source + sig_source[i]/(2*math.pi) * (Y-y_source[i])/((X-x_source[i])**2 + (Y-y_source[i])**2)
    u_sink = u_sink + sig_sink[i]/(2*math.pi) * (X-x_sink[i])/((X-x_sink[i])**2 + (Y-y_sink[i])**2)
    v_sink = v_sink + sig_sink[i]/(2*math.pi) * (Y-y_sink[i])/((X-x_sink[i])**2 + (Y-y_sink[i])**2)
u = u_source + u_sink
v = v_source + v_sink

# plot the streamlines
size = 5
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=1, linewidth=1, arrowsize=2, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
pyplot.scatter(x_sink, y_sink, color='#33cc33', s=80, marker='o', linewidth=0);
```

```python
# sides of a square using a Riemann sum
dx = x[1]-x[0]             # x spacing of grid
dy = y[1]-y[0]             # y spacing of grid

u_int_right = u[:,-1]            # grab last column
v_int_top = v[-1,:]              # grab last row
u_int_left = u[:,1]              # grab first column
v_int_bottom = v[1,:]            # grab first column
Q = numpy.sum(u_int_right*dy) + numpy.sum(v_int_top*dx) - numpy.sum(u_int_left*dy) - \
    numpy.sum(v_int_bottom*dx)        # sum up to approximate integral
print(Q)
```

The computed value is close to the (signed) sum of the source and sink strengths, 2.0, included in the contour of integration.

```python
# Problem 4
N = 50                                # number of points in each direction
x_start, x_end = -4.0, 4.0            # boundaries in the x-direction
y_start, y_end = -4.0, 4.0            # boundaries in the y-direction
x = numpy.linspace(x_start, x_end, N)    # creates a 1D-array with the x-coordinates
y = numpy.linspace(y_start, y_end, N)    # creates a 1D-array with the y-coordinates
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid

# part a
sig_source = 2.0                           # source strength
sig_sink = -2.0                            # sink strength
x_source, y_source = -2.0, 0.0             # location of the source
x_sink, y_sink = 2.0, 0.0                  # location of the sink

# compute the velocity potential on the mesh grid
Phi_source = sig_source/(2*math.pi) * numpy.log(numpy.sqrt((X-x_source)**2 + (Y-y_source)**2))
Phi_sink = sig_sink/(2*math.pi) * numpy.log(numpy.sqrt((X-x_sink)**2 + (Y-y_sink)**2))
Phi = Phi_source + Phi_sink

# plot the potential lines
size = 5
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.contour(X, Y, Phi, levels=numpy.linspace(-1, 1, 20))
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
pyplot.scatter(x_sink, y_sink, color='#33cc33', s=80, marker='o', linewidth=0);
```

```python
# part b
sig_source = 2.0                           # source strength
sig_sink = -1.0                             # sink strength
x_source, y_source = -2.0, 0.0             # location of the source
x_sink, y_sink = 2.0, 0.0                  # location of the sink

# compute the velocity potential on the mesh grid
Phi_source = sig_source/(2*math.pi) * numpy.log(numpy.sqrt((X-x_source)**2 + (Y-y_source)**2))
Phi_sink = sig_sink/(2*math.pi) * numpy.log(numpy.sqrt((X-x_sink)**2 + (Y-y_sink)**2))
Phi = Phi_source + Phi_sink

# plot the potential lines
size = 5
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.contour(X, Y, Phi, levels=numpy.linspace(-1, 1, 20))
pyplot.scatter(x_source, y_source, color='#CD2305', s=80, marker='o', linewidth=0);
pyplot.scatter(x_sink, y_sink, color='#33cc33', s=80, marker='o', linewidth=0);
```
