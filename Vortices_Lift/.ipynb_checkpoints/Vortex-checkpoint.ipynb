```python
import numpy
import math
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams
rcParams['font.size']=16
```

```python
N = 50
x_start,x_end = -2.0,2.0
y_start,y_end = -1.0,1.0
x=numpy.linspace(x_start,x_end,N)
y=numpy.linspace(y_start,y_end,N)
X,Y=numpy.meshgrid(x,y)
```

```python
gamma = 5.0
x_vortex,y_vortex = 0.0,0.0
```

```python
def get_vel_vortex(strength,xv,yv,X,Y):
    u=+strength/(2*numpy.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v=-strength/(2*numpy.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u,v

def get_stream_vortex(strength,xv,yv,X,Y):
    psi = strength/(4*numpy.pi)*numpy.log((X-xv)**2+(Y-yv)**2)
    return psi
```

```python
u_vortex,v_vortex=get_vel_vortex(gamma,x_vortex,y_vortex,X,Y)
psi_vortex=get_stream_vortex(gamma,x_vortex,y_vortex,X,Y)
```

```python
size = 10
pyplot.figure(figsize=(size,(y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,u_vortex,v_vortex,density=2)
pyplot.scatter(x_vortex,y_vortex,color='r',s=80,marker='o');
```

```python
strength_sink=-1.0
x_sink,y_sink=0.0,0.0
```

```python
def velocity_sink(strength,xs,ys,X,Y):
    u=strength/(2*numpy.pi)*(X-xs)/((X-xs)**2+(Y-ys)**2)
    v=strength/(2*numpy.pi)*(Y-ys)/((X-xs)**2+(Y-ys)**2)
    return u,v

def stream_sink(strength,xs,ys,X,Y):
    psi =strength/(2*numpy.pi)*numpy.arctan2((Y-ys),(X-xs))
    return psi
```

```python
u_sink,v_sink= velocity_sink(strength_sink,x_sink,y_sink,X,Y)
psi_sink = stream_sink(strength_sink,x_sink,y_sink,X,Y)
```

```python
u=u_vortex+u_sink
v=v_vortex+v_sink
psi=psi_vortex+psi_sink
```

```python
size=10
pyplot.figure(figsize=(size,(y_end-y_start)/(x_end-x_start)*size))
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,u,v,density=2)
pyplot.scatter(x_vortex,y_vortex,color='r',s=80,marker='o');
```
