```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
N = 201
x_start,x_end = -4.0,4.0
y_start,y_end = -5.0,5.0
x = numpy.linspace(x_start,x_end,N)
y = numpy.linspace(y_start,y_end,N)
X,Y= numpy.meshgrid(x,y)
z = X+1j*Y
```

```python
x_source,y_source = -1.5,0
x_sink,y_sink = 1.5,0
str_source = 0.1
str_sink = -1.0
z_sink = x_sink- 1j*y_sink
z_source = x_source - 1j*y_source
```

```python
W = str_source/(numpy.pi*2)*numpy.log(z-z_source) + str_sink/(numpy.pi*2)*numpy.log(z-z_sink)
```

```python
w =  str_source/(numpy.pi*2)/(z-z_source) + str_sink/(numpy.pi*2)/(z-z_sink)
```

```python
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,w.real,-w.imag,density=7)
```

```python
r = numpy.linspace(0.25,0.5,12)
theta = numpy.linspace(0,2*numpy.pi,15)
R,THETA = numpy.meshgrid(r,theta)
```

```python
zeta = R*numpy.exp(1j*THETA)
```

```python
def j_trans(zeta,a):
    zx = zeta + (a**2/zeta)
    return zx
```

```python
zx = j_trans(zeta,.01)
size=6
pyplot.figure(figsize=(size,size))
pyplot.scatter(zeta.real,zeta.imag)
```

```python
size=6
pyplot.figure(figsize=(size,size))
pyplot.scatter(zx.real,zx.imag)
```
