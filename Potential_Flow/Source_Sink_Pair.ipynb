```python
import math
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
N = 50
x_start, x_end = -2.0,2.0
y_start,y_end = -1.0,1.0
x = numpy.linspace(x_start,x_end,N)
y = numpy.linspace(y_start,y_end,N)

#print('x=',x)
#print('y=',y)

X,Y = numpy.meshgrid(x,y)
```

```python
size =10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x',fontsize=16)
pyplot.ylabel('y',fontsize=16)
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.scatter(X,Y,s=10,color='#895634', marker='o', linewidth=0);
```

```python
strength_source = 5.0
x_source,y_source = -1.0,0.0

u_source = strength_source/(2*math.pi)*(X-x_source)/((X-x_source)**2\
            + (Y-y_source)**2)
v_source = strength_source/(2*math.pi)*(Y-y_source)/((X-x_source)**2\
            + (Y-y_source)**2)
```

```python
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x',fontsize=16)
pyplot.ylabel('y',fontsize=16)
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,u_source,v_source,density=2,linewidth=1,\
                 arrowsize=2, arrowstyle='->')
pyplot.scatter(x_source,y_source, color = '#990000', s=80,\
               marker='o', linewidth=0);
```

```python
strength_sink = -5.0
x_sink,y_sink = 1.0,0.0
u_sink = strength_sink/(2*math.pi)*(X-x_sink)/((X-x_sink)**2+(Y-y_sink)**2)
v_sink = strength_sink/(2*math.pi)*(Y-y_sink)/((X-x_sink)**2+(Y-y_sink)**2)
```

```python
pyplot.figure(figsize=(size,(y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,u_sink,v_sink,density=2,linewidth=1,arrowsize=2,arrowstyle='->')
pyplot.scatter(x_sink,y_sink,color='#990000',s=80,marker='o',linewidth=0);
```

```python
u_pair = u_source + u_sink
v_pair = v_source + v_sink

pyplot.figure(figsize=(size,(y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x')
pyplot.ylabel('y')
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,u_pair,v_pair,density=2.0, linewidth=1,arrowsize=2,arrowstyle='->')
pyplot.scatter([x_source,x_sink],[y_source,y_sink], color='#990000',s=80,marker='o',linewidth=0);
```
