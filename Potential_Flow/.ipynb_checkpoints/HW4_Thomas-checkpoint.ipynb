```python
import numpy
from matplotlib import pyplot
from matplotlib import rcParams
%matplotlib inline
rcParams['font.size']=16
rcParams['font.family']='serif'
```

# 1a

$$\psi = u_{\infty}rsin(\theta)-\frac{ksin(\theta)}{2\pi r}$$

Set $\psi=0$:

$$u_{\infty}rsin(\theta) = \frac{ksin(\theta)}{2\pi r}$$

$$r^{2} = \frac{k}{u_{\infty}2\pi}$$

$$r = \sqrt{\frac{k}{u_{\infty}2\pi}}$$

# 1b

```python
k = 1
u_inf = 1
r = numpy.sqrt(k/(2*numpy.pi*u_inf))
```

```python
#print(r)
theta=numpy.linspace(0,2*numpy.pi,100)
x_r = r*numpy.cos(theta)
y_r=r*numpy.sin(theta)
```

```python
N = 200
y_start,y_end = -1.5,1.5
x_start,x_end = -2.0,2.0
x=numpy.linspace(x_start,x_end,N)
y = numpy.linspace(y_start,y_end,N)
X,Y=numpy.meshgrid(x,y)
```

```python
def get_vel_doublet(strength,xd,yd,X,Y):
    u = -strength/(2*numpy.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = -strength/(2*numpy.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    return u,v

def get_stream_doublet(strength,xd,yd,X,Y):
    psi = -strength/(2*numpy.pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
    return psi
```

```python
x_doublet,y_doublet = 0,0
u_doublet,v_doublet=get_vel_doublet(k,x_doublet,y_doublet,X,Y)
psi_doublet = get_stream_doublet(k,x_doublet,y_doublet,X,Y)

u_fs = u_inf*numpy.ones((N,N),dtype=float)
v_fs = numpy.zeros((N,N),dtype=float)
psi_fs = u_inf*y

u = u_fs+u_doublet
v = v_fs+v_doublet
psi = psi_doublet+psi_fs
```

```python
size = 10
pyplot.figure(figsize=(size,(y_end-y_start)/(x_end-x_start)*size))
pyplot.xlim(x_start,x_end)
pyplot.ylim(y_start,y_end)
pyplot.streamplot(X,Y,u,v,density =2)
pyplot.scatter(x_r,y_r,marker='.',color='#500000');
```

# 2a

From the *Doublet* Notebook, we are given

$$C_{p}= 1 - ( \frac{U}{U_\infty} )^{2}$$

First find velocities from stream function:

$$V_{r} = \frac{1}{r} \frac{\partial \psi}{\partial \theta} = U_{\infty}cos \theta(1- \frac{R^{2}}{r^{2}})$$

$$V_{\theta} = - \frac{\partial \psi}{\partial r} = -U_{\infty}sin \theta (1+ \frac {R^{2}}{r^{2}})$$

We want to find $C_{p}$ on the surface of the cylinder, so in the velocity equations, set r=R:

$$V_{r} = 0 $$

$$V_{\theta} = -2 U_{\infty} sin \theta $$

We can now plug in the velocity at the surface of the cylinder into the $C_{p}$ equation, which is just $V_{\theta}$:

$$C_{p} = 1 - ( \frac{-2 U_{\infty} sin \theta}{U_{\infty}} )^{2}$$

$$C_{p} = 1 - 4 sin^{2} \theta$$

# 2b

```python
rr=numpy.linspace(0,1.5,N)
theta = numpy.linspace(0,2*numpy.pi,N)
R,THETA = numpy.meshgrid(rr,theta)
rm=r*numpy.ones_like(theta)
```

```python
ax=pyplot.subplot(111,polar=True)
ax.plot(THETA,R,color='r')
ax.set_rmax(1.5)
ax.grid(True);
```

```python
print(rm[1])#radius of circle
print(rr[53])#closest r value to actual radius. to improve increase N, number of time steps
```

```python
def get_vel_cyl(u_inf,R,THETA):
    v_r = u_inf*numpy.cos(THETA)*(1-(r**2/(R)**2))
    v_theta = -u_inf*numpy.sin(THETA)*(1+(r**2/(R)**2))
    return v_r,v_theta

def get_stream_cyl(u_inf,R,THETA):
    psi = u_inf*(R)*numpy.sin(THETA)*(1-(r**2/(R)**2))
    return psi
```

```python
v_r,v_theta = get_vel_cyl(u_inf,R,THETA)
psi_doub = get_stream_cyl(u_inf,R,THETA)
```

```python
cp = 1.0-(v_r**2+v_theta**2)/u_inf**2

ax=pyplot.subplot(111,polar=True)
contf = pyplot.contourf(THETA,R,cp,levels=numpy.linspace(-2.0,1.0,100),extend='both')
cbar=pyplot.colorbar(contf)
cbar.set_label('$C_p$')
cbar.set_ticks([-2.0,-1.0,0.0,1.0])
pyplot.scatter(theta,rm,marker='.');
```

```python
print(cp[:,53])# cp values for r=0.399 (closest value to radius of circle, r=0.3989)
```

```python
cp_plot = cp[:,53]
```

```python
cp_theo = 1-(4*(numpy.sin(theta))**2)
```

```python
pyplot.figure(figsize=(size,size/2))
pyplot.plot(theta,cp_theo,linewidth=1,label='Theoretical',marker='o')
pyplot.plot(theta,cp_plot,marker='.',label='Plotted',linewidth=1.5)
pyplot.xlim(0,2*numpy.pi)
pyplot.legend();
```

The theoretical and actual values of $C_{p}$ appear to be exactly the same. However, we know that the plotted values are slightly off since we did not get the $C_{p}$ values at the exact radius (we did not have a radius of 0.3989 on our mesh grid). When printing the values for each (below), they tend to be within 2 decimal places of each other.
