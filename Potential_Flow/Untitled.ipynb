```python
import numpy
from matplotlib import pyplot


r = numpy.linspace(0.25,0.5,12)
theta = numpy.linspace(0,2*numpy.pi,15)
R,THETA = numpy.meshgrid(r,theta)

zeta = R*numpy.exp(1j*THETA)
```

```python
def jt(zeta,a):
    z = zeta + a**2/zeta
    return z
```

```python
z = jt(R*numpy.exp(1j*THETA),0.23)
#print (z)
```

```python
size=6
pyplot.figure(figsize=(size,size))
pyplot.scatter(zeta.real,zeta.imag)
```

```python
pyplot.figure(figsize=(size,size))
pyplot.scatter(z.real,z.imag)
```
